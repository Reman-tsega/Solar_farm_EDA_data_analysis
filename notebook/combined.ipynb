# combined data EDA

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# # # Load the datasets

```python

benin_data = pd.read_csv('./../data/benin-malanville.csv')
sierra_leone_data = pd.read_csv('./../data/sierraleone-bumbuna.csv')
togo_data = pd.read_csv('./../data/togo-dapaong_qc.csv')
```

# # # Combine the datasets for comprehensive analysis


```python
combined_data = pd.concat([benin_data, sierra_leone_data, togo_data], ignore_index=True)
```

# # # # outliers 

```python
outliers = combined_data[(combined_data['GHI'] < 0) | (combined_data['DNI'] < 0) | (combined_data['DHI'] < 0)]
# Boxplots to detect outliers
plt.figure(figsize=(15, 10))

# Boxplot for GHI, DNI, DHI
plt.subplot(2, 3, 1)
sns.boxplot(x=combined_data['GHI'])
plt.title('Boxplot for GHI')

plt.subplot(2, 3, 2)
sns.boxplot(x=combined_data['DNI'])
plt.title('Boxplot for DNI')

plt.subplot(2, 3, 3)
sns.boxplot(x=combined_data['DHI'])
plt.title('Boxplot for DHI')

# Boxplot for ModA, ModB
plt.subplot(2, 3, 4)
sns.boxplot(x=combined_data['ModA'])
plt.title('Boxplot for ModA')

plt.subplot(2, 3, 5)
sns.boxplot(x=combined_data['ModB'])
plt.title('Boxplot for ModB')

# Boxplot for WS and WSgust
plt.subplot(2, 3, 6)
sns.boxplot(x=combined_data['WSgust'])
plt.title('Boxplot for WSgust')

plt.tight_layout()
plt.show()
```

# # # Time Series Analysis

```python
combined_data.columns
```

```python
# Convert the Timestamp to datetime format
combined_data['Timestamp'] = pd.to_datetime(combined_data['Timestamp'])

# Set the Timestamp as the index
combined_data.set_index('Timestamp', inplace=True)

# Plotting GHI, DNI, DHI, and Tamb over time
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
combined_data['GHI'].plot()
plt.title('GHI Over Time')
plt.show()

```

```python

plt.subplot(2, 2, 2)
combined_data['DNI'].plot()
plt.title('DNI Over Time')
plt.show()
```

```python
plt.subplot(2, 2, 3)
combined_data['DHI'].plot()
plt.title('DHI Over Time')
plt.show()
```

```python
plt.subplot(2, 2, 4)
combined_data['Tamb'].plot()
plt.title('Tamb Over Time')
plt.show()
```

## Step 4: Evaluate the Impact of Cleaning on Sensor Readings

```python
# Group by Cleaning status and calculate the mean of ModA and ModB
cleaning_effect = combined_data.groupby('Cleaning')[['ModA', 'ModB']].mean()
print(cleaning_effect)

# Plot the impact of cleaning
cleaning_effect.plot(kind='bar', figsize=(10, 6))
plt.title('Impact of Cleaning on ModA and ModB')
plt.show()
```

# # # Correlation Analysis


```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
corr_matrix = combined_data[['GHI', 'DNI', 'DHI', 'TModA', 'TModB', 'WS', 'WSgust', 'WD']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
